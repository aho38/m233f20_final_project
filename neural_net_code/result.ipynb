```python
import pickle
import numpy as np
import math
import matplotlib.pyplot as plt
import torch
#from test_data_gen import zero_isocontour
from IPython.display import Image
```

## Circular Interface

```python
x = pickle.load(open('./result/circular/exp3tensor_y.pkl', 'rb'))
y = pickle.load(open('./result/circular/exp3tensor_y_hat.pkl', 'rb'))
p = np.polyfit(x.squeeze(),y.squeeze(),1)

fig = plt.figure(figsize=(8,8))
plt.plot([x.min(), x.max()], [x.min()* p[0] + p[1], x.max() * p[0] + p[1]], 'r')
plt.plot(x,x, '--', c='k')
plt.scatter(x,y,s=50,facecolor='none', edgecolor='b')
plt.legend(('fit', 'predicted=expected', 'data'))
plt.grid(linestyle='--')
plt.xlabel('Target hk')
plt.ylabel('Approximated hk')
plt.title('Circular Interface')
```

## Reinitialized three pedal 5

```python
x = pickle.load(open('./result/petal/5_reinit_tensor_y.pkl', 'rb'))
y = pickle.load(open('./result/petal/5_reinit_tensor_y_hat.pkl', 'rb'))
p = np.polyfit(x.squeeze(),y.squeeze(),1)
print(p)

fig = plt.figure(figsize=(8,8))
plt.plot([x.min(), x.max()], [x.min()* p[0] + p[1], x.max() * p[0] + p[1]], 'r')
plt.plot(x,x, '--', c='k')
plt.scatter(x,y,s=50,facecolor='none', edgecolor='b')
plt.legend(('fit', 'predicted=expected', 'data'))
plt.ylim([x.min()-0.005,x.max()+0.005])
plt.xlim([x.min()-0.005,x.max()+0.005])
plt.grid(linestyle='--')
plt.xlabel('Target hk')
plt.ylabel('Approximated hk')
plt.title('5 Reinitialization Three Petal')
```
